```python
import matplotlib.pyplot as plt
import numpy as np

# Provided data
labels = ['Python', 'Java']

# Time for each sorting algorithm
merge_sort_times = [890195.77, 497249]
quick_sort_times = [735394.03, 483405]

x = np.arange(len(labels))  # Label locations
width = 0.35  # Width of the bars

plt.figure(figsize=(10, 6))

# Plot bars
plt.bar(x - width/2, merge_sort_times, width, label='Merge Sort', color='skyblue', edgecolor='black')
plt.bar(x + width/2, quick_sort_times, width, label='Quick Sort', color='orange', edgecolor='black')

# Labels and title
plt.xlabel('Language')
plt.ylabel('Time (ms)')
plt.title('Merge Sort vs Quick Sort Time Comparison(Python vs Java)')
plt.xticks(x, labels)
plt.legend()

# Add data labels on top of bars
for i in range(len(labels)):
    plt.text(x[i] - width/2, merge_sort_times[i] + max(merge_sort_times) * 0.01, f"{merge_sort_times[i]:.2f}", ha='center', va='bottom')
    plt.text(x[i] + width/2, quick_sort_times[i] + max(merge_sort_times) * 0.01, f"{quick_sort_times[i]:.2f}", ha='center', va='bottom')

plt.ylim(0, max(max(merge_sort_times), max(quick_sort_times)) + 100000)  # Adjust Y limit
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
cases = ['Best Case', 'Average Case', 'Worst Case']
python_times = [0.00200, 0.01640, 0.01590]
java_times = [0.00050, 0.00640, 0.00940]

# Position of bars on X-axis
x = np.arange(len(cases))
width = 0.35  # Width of the bars

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(x - width/2, python_times, width, label='Python', color='skyblue', edgecolor='black')
plt.bar(x + width/2, java_times, width, label='Java', color='orange', edgecolor='black')

# Labels and title
plt.xlabel('Case')
plt.ylabel('Time (ms)')
plt.title('Binary Search Time Comparison(Python vs Java)')
plt.xticks(x, cases)
plt.legend()

# Add data labels on top of bars
for i in range(len(cases)):
    plt.text(x[i] - width/2, python_times[i] + 0.0005, f"{python_times[i]:.5f}", ha='center', va='bottom', fontsize=9)
    plt.text(x[i] + width/2, java_times[i] + 0.0005, f"{java_times[i]:.5f}", ha='center', va='bottom', fontsize=9)

plt.ylim(0, max(max(python_times), max(java_times)) + 0.005)  # Adjust y-axis limit for label spacing
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
